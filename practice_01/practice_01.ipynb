```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
# Câu 1:Tạo DataFrame df chứa danh sách các sinh viên khoa CN, khoa TO và khoa VL. 
#Các danh sách này được đọc từ các tập tin sinh_vien_CN.csv, sinh_vien_TO.csv và sinh_vien_VL.csv
khoa_cn = pd.read_csv('sinh_vien_CN.csv')
khoa_to = pd.read_csv('sinh_vien_TO.csv')
khoa_vl = pd.read_csv('sinh_vien_VL.csv')
df = pd.concat([khoa_cn,khoa_to,khoa_vl])
```

```python
# Câu 4: Tạo DataFrame df_ket_qua từ tập tin ket_qua.csv
df_ket_qua = pd.read_csv('ket_qua.csv')
df_ket_qua
```

```python
# Câu 5: Tạo DataFrame df_dtb bằng cách nhóm df_ket_qua theo masv và tính điểm trung bình cho mỗi sinh viên.
df_dtb = df_ket_qua.groupby('masv', as_index=False).diem.mean()
df_dtb
```

```python
# Câu 6: Trộn df và df_dtb theo masv và có kết quả của df như sau:
df = df.merge(df_dtb, on='masv')
df
```

```python
# Câu 7: In ra các sinh viên của khoa CN và có diem>8
diem_CN = df[(df['makh'] == 'CN') & (df['diem'] > 8) ]
diem_CN
```

```python
# Câu 8: In ra các sinh viên của khoa CN và có diem>8. Sắp xếp tăng theo ten và giảm theo diem
diem_CN.sort_values(by='diem', ascending=False)
```

```python
# Câu 9: In ra các nam sinh viên không có học bổng
nam_sv_khong_hb = df[ (df['gioitinh'] == True) & (df['hocbong'] == 0) ]
nam_sv_khong_hb
```

```python
# Câu 10: In ra tổng học bổng của mỗi khoa
hoc_bong_khoa = df.groupby('makh', as_index=False).hocbong.sum()
hoc_bong_khoa
```

```python
# Câu 11: Thống kê học bổng của mỗi khoa.
thong_ke_hoc_bong = df.groupby('makh', as_index=False).hocbong.agg(['sum','min','max','mean'])
thong_ke_hoc_bong
```

```python
# Câu 12: Thống kê điểm của mỗi khoa.
thong_ke_diem_khoa = df.groupby('makh', as_index=False).diem.agg(['count','max','min','mean'])
thong_ke_diem_khoa
```

```python
# Câu 13: Cho biết các sinh viên có điểm lớn nhất, điểm nhỏ nhất
max_diem = df.sort_values(by='diem', ascending=False)
max_diem.head(1)
```

```python
max_diem.tail(1)
```

```python
# Câu 14: Trong DataFrame df tạo thêm cột ketqua với kết quả là giỏi nếu điểm>=8 và kết quả là khá nếu điểm<8.
df['ketqua'] = df['diem'].map(lambda x: 'Gioi' if x >= 8 else 'Kha')
df
```

```python
# Câu 15: Thống kê theo mẫu:
thong_ke = df.groupby(['makh','ketqua'], as_index=False).masv.count()
thong_ke
```
