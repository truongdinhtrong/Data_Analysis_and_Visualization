```python
#Câu: 3	Tạo các DataFrame departments chứa danh sách các nhân viên và jobs chứa danh sách các công việc. Các danh sách này được đọc từ các tập tin departments.xls và job.xls
```

```python
departments = pd.read_excel('departments.xls')
departments
```

```python
jobs = pd.read_excel('jobs.xls')
jobs
```

```python
#Câu 4: 4	Tạo DataFrame emp là danh sách các nhân viên của DataFrame employees chỉ với các cột theo mẫu sau:
```

```python
emp = employees[['EMPLOYEE_ID','FIRST_NAME','LAST_NAME','HIRE_DATE','SALARY','DEPARTMENT_ID','JOB_ID']]
emp.head()
```

```python
# Câu: 5	Trong DataFrame emp hãy thêm vào hai cột: DEPARTMENT_NAME (tên phòng) và JOB_TITLE (tên công việc)  theo mẫu sau:
```

```python
emp = pd.merge(emp,departments[['DEPARTMENT_NAME','DEPARTMENT_ID']],how='left')
```

```python
emp = pd.merge(emp,jobs[['JOB_TITLE','JOB_ID']],how='left')
emp.head()
```

```python
#Câu 6: Liệt kê các nhân viên của phòng có department_id là 50, sắp giảm theo salary
```

```python
emp[emp['DEPARTMENT_ID'] == 50.0].sort_values(by='SALARY',ascending=False).head(10)
```

```python
#Câu: 7	Liệt kê các nhân viên của phòng có department_name là Sales và có salary>=5000 và salary<=7000
```

```python
emp[ (emp['DEPARTMENT_NAME'] == 'Sales') & ( (emp['SALARY'] >= 5000 ) & (emp['SALARY'] <= 7000 )) ]
```

```python
#Câu: 8	Liệt kê các nhân viên làm các công việc sao cho mã công việc (job_id) có chứa chuỗi MAN
```

```python
emp[emp['JOB_ID'].str.contains("MAN")]
```

```python
#Câu: 9	Cho biết nhân viên không thuộc phòng ban nào cả
```

```python
emp[emp['DEPARTMENT_ID'].isnull()]
```

```python
#Câu: 10	In thông tin thống kê của salary
```

```python
emp['SALARY'].describe()
```

```python
#11	Liệt kê các nhân viên có lương (salary) > trung bình các lương, sắp tăng theo department_name và sắp giảm theo salary
```

```python
emp [emp['SALARY'] > emp['SALARY'].mean()].sort_values(by=['DEPARTMENT_NAME','SALARY'], ascending=(True,False))
```

```python
# CÂu: 12	Cho biết nhân viên có lương lớn nhất, nhân viên có lương nhỏ nhất
```

```python
min_salary, max_salary = emp['SALARY'].min(), emp['SALARY'].max()
print('employee min salary:',min_salary)
print('employee max salary:',max_salary)
```

```python
#13	Tạo cột BONUS có giá trị là SALARY + 500 và in ra 5 nhân viên đầu tiên theo mẫu sau:
emp['BONUS'] = emp['SALARY'] + 500
emp[['EMPLOYEE_ID','FIRST_NAME','SALARY','BONUS']].head()
```

```python
# Câu: 14	Tạo cột NAM_VAO_LAM là năm vào làm (trích năm từ cột HIRE_DATE). Xóa cột HIRE_DATE
emp['NAM_VAO_LAM'] = emp['HIRE_DATE'].str.slice(0,4)
emp
```

```python
del emp['HIRE_DATE']
```

```python
emp[['EMPLOYEE_ID','FIRST_NAME','NAM_VAO_LAM']].head()
```

```python
# Câu: 15	Thống kê theo mẫu sau:
```

```python
emp.groupby('DEPARTMENT_NAME',as_index=False).SALARY.agg(['count','max','min','mean','sum']).reset_index()
```

```python
#Câu: 16	Thống kê theo mẫu sau:
emp.groupby('NAM_VAO_LAM').size().reset_index(name='COUNT').sort_values(by='COUNT',ascending=False)
```

```python
#câu: 17	Thống kê theo mẫu sau:
emp.groupby(['DEPARTMENT_NAME','JOB_TITLE']).size().reset_index(name='COUNT').head(11)
```

```python
#18	Vẽ biểu đồ theo mẫu sau: (đếm số nhân viên theo phòng)
sb.countplot(data=emp, x='DEPARTMENT_ID')
```

```python
#Câu: 19	Vẽ biểu đồ giống như câu 18 nhưng thay các departmnet_id (trục X) bằng các department_name (có xoay 45 độ)
ax = sb.countplot(data=emp, x='DEPARTMENT_NAME')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
#Câu: 20	Vẽ biểu đồ theo mẫu sau: (vẽ boxplot theo lương tất cả nhân viên)
```

```python
#Câu: 21	Vẽ biểu đồ theo mẫu sau: (vẽ boxplot theo lương các nhân viên của phòng 80)
```

```python
#Câu 22	Vẽ biểu đồ theo mẫu sau: (vẽ boxplot cho tất cả các phòng và theo lương)
```

```python
sb.boxplot(data=emp, x='DEPARTMENT_ID', y='SALARY');
```

```python
#Câu 23	Vẽ biểu đồ theo mẫu sau: (vẽ barplot với tổng lương theo phòng)
data_d = emp.groupby('DEPARTMENT_NAME').SALARY.sum().reset_index(name='SALARY')
data_d
sb.barplot(data=data_d, y='DEPARTMENT_NAME',x='SALARY');
```

```python
#Câu 24	Vẽ biểu đồ theo mẫu sau: (vẽ barplot với số nhân viên vào làm theo năm)
#sb.barplot(data=data_n, x='NAM_VAO_LAM',y='EMPLOYEE_ID')
data_n = emp.groupby('NAM_VAO_LAM').size().reset_index(name='COUNT')
sb.barplot(data=data_n, x='NAM_VAO_LAM',y='COUNT');
```

```python
#Câu: 25	Vẽ biểu đồ giống như câu 24 nhưng vẽ với số nhân viên vào làm theo năm giảm dần (tính từ trái qua phải)

sb.barplot(data=data_n, x='NAM_VAO_LAM',y='COUNT', order = emp['NAM_VAO_LAM'].value_counts(ascending=False).index);

```

```python
#26	Vẽ biểu đồ theo mẫu sau: (chỉ vẽ với hai phòng và vẽ theo tổng lương)
tong_luong_sales_shipping = emp.groupby(['DEPARTMENT_NAME']).SALARY.sum()
data_sales_shipping= tong_luong_sales_shipping[['Sales','Shipping']].reset_index()
```

```python
sb.catplot(data=data_sales_shipping, col='DEPARTMENT_NAME', x='SALARY', kind='bar');
```
